```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("../data/train_preprocessed.csv")
train_x = train.drop(["target"], axis=1)
train_y = train["target"]

test_x = pd.read_csv("../data/test_preprocessed.csv")
test_x
```

```python
from sklearn.metrics import log_loss
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=4, shuffle=True, random_state=71)
tr_idx, va_idx = list(kf.split(train_x))[0]

tr_x, va_x = train_x.iloc[tr_idx], train_x.iloc[va_idx]
tr_y, va_y = train_y.iloc[tr_idx], train_y.iloc[va_idx]
```

```python
from models import Model2Linear, ModleGBDT
```

```python
model_1a = ModleGBDT()
model_1a.fit(tr_x, tr_y, va_x, va_y)
va_pred_1a = model_1a.predict(va_x)
test_pred_1a = model_1a.predict(test_x)
```

```python
print(f'logloss: {log_loss(va_y, va_pred_1a, eps=1e-7):.4f}')
```

```python
va_x_2 = pd.DataFrame({"pred_1a":va_pred_1a})
test_x_2 = pd.DataFrame({"pred_1a":test_pred_1a})
```

```python
model2 = Model2Linear()
model2.fit(va_x_2, va_y, None, None)

pred_test_2 = model2.predict(test_x_2)
```
