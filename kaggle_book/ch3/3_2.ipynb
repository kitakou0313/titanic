```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('../data/train.csv')
train_x = train.drop(["target"], axis=1)
train_y = train["target"]
test_x = pd.read_csv("../data/test.csv")
```

```python
train
```

```python
train_x_saved = train_x.copy()
test_x_saved = test_x.copy()

def load_data():
    """
    オリジナルのdataframeを返す
    """
    return train_x_saved.copy(), test_x_saved.copy()
```

```python
cat_cols = ['sex', 'product', 'medical_info_b2', 'medical_info_b3']
```

```python
# One-hot encoding
train_x, test_x = load_data()


import category_encoders as ce

ce_ohe = ce.OneHotEncoder(cols=cat_cols)
train_x = ce_ohe.fit_transform(train_x)
test_x = ce_ohe.transform(test_x)

train_x
```

```python
# label encoding
train_x, test_x = load_data()
```

```python
from sklearn.preprocessing import LabelEncoder

for c in cat_cols:
    le = LabelEncoder()
    le.fit(train_x[c])

    train_x[c] = le.transform(train_x[c])
    test_x[c] = le.transform(test_x[c])

train_x
```

```python
# frequency encoding
train_x, test_x = load_data()
```

```python
for c in cat_cols:
    freq = train_x[c].value_counts()
    print(freq)
    train_x[c] = train_x[c].map(freq)
    test_x[c] = test_x[c].map(freq)

train_x
```

```python
# target encodering
train_x, test_x = load_data()

ce = ce.CatBoostEncoder()
for c in cat_cols:
    train_x[c] = ce.fit_transform(train_x[c], train_y)
    test_x[c] = ce.transform(test_x[c])

train_x
```
