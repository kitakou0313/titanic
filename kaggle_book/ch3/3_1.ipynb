```python
# 対数変換
x = np.array(
    [1.0, 10.0, 100.0, 1000.0, 10000.0]
)

x_loged = np.log10(x)
# xに0が含まれる時用
x1 = np.log1p(x)
# xに負の値が含まれる時用
x2 = np.sign(x) * np.log1p(np.abs(x))

x1, x2, x_loged
```

```python
#Box-Cox変換
# 対数変換の一般化
# できるだけ正規分布に近づくように変換してくれる
train_x, test_x = load_data()
```

### Box-Cox変換
- x^λ = 
    - (x^λ - 1)/λ {λ != 0}
    - log x {λ == 0}
    - ラムダは自動調整してくれる

```python
pos_cols = [c for c in num_cols if (train_x[c] > 0.0).all() and (test_x[c] > 0.0).all()]
pos_cols
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method="box-cox")
pt.fit(train_x[pos_cols])

train_x[pos_cols] = pt.transform(train_x[pos_cols])
test_x[pos_cols] = pt.transform(test_x[pos_cols])

train_x
```

```python
# Yeo-Johnson変換
# Box-Cox変換の一般化

pt_yeo_johnson = PowerTransformer(method="yeo-johnson")
pt_yeo_johnson.fit(train_x[num_cols])

train_x[num_cols] = pt_yeo_johnson.transform(train_x[num_cols])
test_x[num_cols] = pt_yeo_johnson.transform(test_x[num_cols])
```

### Cliping
- データの外れ値を取り除く
- pandas.Dataframeでは境界値に設定される

```python
train_x, test_x = load_data()

p01 = train_x[num_cols].quantile(0.01)
p99 = train_x[num_cols].quantile(0.99)
```

```python
train_x[num_cols] = train_x[num_cols].clip(p01, p99, axis=1)
train_x
```

### RankGauss
- 順序づけ -> 正規分布に変換
- NNに対しては通常の標準化よりよい成績が出るとのこと

```python
from sklearn.preprocessing import QuantileTransformer

train_x, test_x = load_data()

transformer = QuantileTransformer(n_quantiles=100, random_state=0, output_distribution="normal")
transformer.fit(train_x[num_cols])

train_x[num_cols] = transformer.transform(train_x[num_cols])
```
